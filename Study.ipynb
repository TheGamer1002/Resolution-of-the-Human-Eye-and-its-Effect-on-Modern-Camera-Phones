```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import itertools

# Load data from CSV file
data = pd.read_csv('/Users/parker/Documents/data.csv')
tstats = []
# Define the order of the photos
photo_order = [1, 2, 3, 4, 5, 6]
stuff = [data["photo_1"],data["photo_2"],data["photo_3"],data["photo_4"],data["photo_5"],data["photo_6"]]
for L in range(len(stuff) + 1):
    for subset in itertools.combinations(stuff, L):
        tstats += ttest_ind(L, subset)
means = data.mean()
stds = data.std()
photo_1 = data['photo_1']
photo_2 = data['photo_2']
t_stat, p_val = ttest_ind(photo_1, photo_2)
# Create a color map for the lines
color_map = plt.get_cmap('viridis')

# Define the plot dimensions and style
fig, ax = plt.subplots(figsize=(10, 6)) 



plt.style.use('ggplot')

# Loop over each completed test and plot a line
for i, row in data.iterrows():
    # Get the values for each photo in the order they were ranked
    photo_values = [row[f'photo_{j}'] for j in photo_order]
    # Plot a line connecting the values for each photo
    ax.plot(photo_order, photo_values, color=color_map(i / len(data)), alpha=0.8)

# Add reference lines for the correct photo order
correct_order = [1, 2, 3, 4, 5, 6]
ax.plot(photo_order, correct_order, color='black', alpha=0.8, label='Correct Order')

# Set the plot title, axis labels, and legend
ax.set_title('Participant Rankings of Photos')
ax.set_xlabel('Photo Number')
ax.set_ylabel('Ranking')
ax.legend()
plt.savefig('e2.png')
# Show the plot
plt.show()
plt.savefig('e.png')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from itertools import permutations
from scipy.stats import ttest_ind

# Load data
data = pd.read_csv('Documents/data.csv')

# Create all possible pairs of photos
photo_pairs = list(permutations(range(1, 7), 2))

# Calculate t-tests for all pairs of photos
ttest_results = {}
for pair in photo_pairs:
    photo1 = data[f'photo_{pair[0]}']
    photo2 = data[f'photo_{pair[1]}']
    t, p = ttest_ind(photo1, photo2)
    ttest_results[pair] = {'t': t, 'p': p}

# Create heatmap of p-values
p_values = pd.DataFrame.from_dict(ttest_results, orient='index')['p']
p_values = p_values.unstack().fillna(1)
fig, (ax2) = plt.subplots(ncols=1, figsize=(12, 6))


# Create table of t-tests


p_values.index.name = 'Photo 1'
p_values.columns.name = 'Photo 2'
p_values = p_values.applymap(lambda x: f'{x:.2f}' if x else '')




table2 = ax2.table(cellText=p_values.values, rowLabels=p_values.index, colLabels=p_values.columns, loc='center')
table2.auto_set_font_size(False)
table2.set_fontsize(14)
ax2.axis('off')
ax2.set_title('T-Test Results')
plt.savefig('e3.png')
print(ttest_results)
plt.show()
```

```python
### bonus graph!

import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d import axes3d
from matplotlib import cm
from scipy import interpolate
%matplotlib inline


array_distribution3d = np.array([[2, 1, 0, 1, 2, 1],
               [1, 1, 0, 2, 0, 3],
               [1, 0, 2, 1, 2, 1],
               [2, 1, 1, 1, 2, 0],
               [0, 3, 2, 1, 1, 0],
               [1, 1, 2, 1, 0, 2]])
# create the meshgrid to plot on
x = np.arange(0, array_distribution3d.shape[0])
y = np.arange(0, array_distribution3d.shape[1])

# here are the x,y and respective z values
X, Y = np.meshgrid(x, y)
z = []
for i in range(0, array_distribution3d.shape[0]):
    z_y = []
    for j in range(0, array_distribution3d.shape[1]):
        z_y.append(array_distribution3d[i, j])

    z.append(z_y)
Z = np.array(z)

# create the figure, add a 3d axis, set the viewing angle
fig = plt.figure()
ax = fig.add_subplot( projection='3d')


# here we create the surface plot
ax.plot_surface(X, Y, Z, edgecolor='royalblue', lw=0.5,
                alpha=0.3)


ax.contour(X, Y, Z, zdir='z', offset=-1, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='x', offset=6, cmap='coolwarm')
ax.contour(X, Y, Z, zdir='y', offset=6, cmap='coolwarm')

ax.set(xlim=(0, 6), ylim=(0, 6), zlim=(-1, 3),
       xlabel='X', ylabel='Y', zlabel='Z')
ax.invert_xaxis()
```
